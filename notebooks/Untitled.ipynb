```python
import pandas as pd
```

```python
covid_time = pd.read_csv('./../src/datasets/covid_19_data.csv')
```

```python
covid_time.head()
```

```python
all_dates = covid_time['ObservationDate'].unique()
all_countries = covid_time['Country/Region'].unique()
```

```python
country_time_data = {}
```

```python
for date in all_dates:
    country_time_data[date]={}
    for country in all_countries:
        country_time_data[date][country] = {}
        country_time_data[date][country]['confirmed'] = covid_time.loc[(covid_time['ObservationDate'] == date) & (covid_time['Country/Region'] == country), 'Confirmed'].sum()
        country_time_data[date][country]['deaths'] = covid_time.loc[(covid_time['ObservationDate'] == date) & (covid_time['Country/Region'] == country), 'Deaths'].sum()
        country_time_data[date][country]['recovered'] = covid_time.loc[(covid_time['ObservationDate'] == date) & (covid_time['Country/Region'] == country), 'Recovered'].sum()
```

```python
country_time_data
```

```python
import json
country_time_json = json.dumps(country_time_data) 
text_file = open("../src/datasets/country_time.json", "w")
text_file.write(country_time_json)
text_file.close()
```
